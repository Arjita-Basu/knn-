**Accuracy score of KNN model**

```python
from sklearn.model_selection import train_test_split
X = np.array(df.iloc[:, 0:4])
y = np.array(df['class'])
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
predict=knn.predict(x_test)
predict
```

```python
neighbors = list(range(1,50,2))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predict))
```

**CHecking for optimal value of K**

**Applying k-fold cross validation**

```python
cv_scores=[]
import sklearn.model_selection
from sklearn.model_selection import cross_val_score
```

```python
for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores= sklearn.model_selection.cross_val_score(knn,x_train,y_train,cv=10)
    cv_scores.append(scores.mean())
```

**MSE(mean squared error) calculation**

```python
MSE=[1-x for x in cv_scores]
print("Minimum MSE is ",min(MSE))
optimal_k = neighbors[MSE.index(min(MSE))]
print ("The optimal number of neighbors are ",optimal_k)

# plot misclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel('Number of KNeighbors')
plt.ylabel('Misclassification Error')
plt.show()
```

```python
K=[]
for x in range(1,26):
    y=1/x
    K.append(y)
```

**Plot between MSE and 1/K**

```python
plt.plot(K, MSE)
plt.xlabel('1/K')
plt.ylabel('MSE')
plt.show()
```
